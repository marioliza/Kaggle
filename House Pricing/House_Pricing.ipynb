```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore

# Define the threshold for outliers
z_threshold = 2.5

# List of numerical columns excluding 'SalePrice'
numerical_features = train_df.select_dtypes(include=[np.number]).columns.drop('SalePrice')

# Loop through each numerical feature
for feature in numerical_features:
    # Calculate the z-scores for the feature with respect to SalePrice
    train_df['z_score'] = zscore(train_df[feature])

    # Identify outliers
    outliers = train_df[np.abs(train_df['z_score']) > z_threshold]

    # Plot the feature vs SalePrice, highlighting outliers
    plt.figure(figsize=(10, 6))
    plt.scatter(train_df[feature], train_df['SalePrice'], label='Data', alpha=0.5)
    plt.scatter(outliers[feature], outliers['SalePrice'], color='red', label='Outliers', alpha=0.7)

    # Add labels and title
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f'SalePrice vs {feature} with Outliers Highlighted')
    plt.legend()
    plt.show()

# Drop the temporary 'z_score' column after plotting
train_df.drop(columns=['z_score'], inplace=True)
```
